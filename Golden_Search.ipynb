```python
import numpy as np
import matplotlib.pyplot as plt

def golden_search(f, ip, ep, tol=1e-5):

  # Step 2: Intermediate points
  p1 = ip + rho * (ep - ip)
  p2 = ip + (1 - rho) * (ep - ip)

  fp1 = f(p1)
  fp2 = f(p2)

  tol= 1e-2

  while abs(ep - ip) > tol:
  #for i in range(1,5):

          if fp1 < fp2:
              ep = p2
              p2 = p1
              p1 = ip + rho * (ep - ip)
              fp2 = fp1
              fp1 = f(p1)
              print(f"Reduced interval = [{ip}, {ep}], ")
          else:
              ip = p1
              p1 = p2
              p2 = ip + (1 - rho) * (ep - ip)
              fp1 = fp2
              fp2 = f(p2)
              print(f"Reduced interval = [{ip}, {ep}]")
  xmin = (ip+ep)/2
  return xmin

ip=0
ep=2
rho =(3-np.sqrt(5))/2 #(np.sqrt(5) - 1) / 2
xmin1= golden_search(f, ip, ep, tol=1e-5)
print(f"Minimum point is {xmin1} and minimizer is {f(xmin1)}" )
```

```python
# Plotting the function
x_vals = np.linspace(ip, ep, 400)
y_vals = f(x_vals)

plt.plot(x_vals, y_vals, label=r'$f(x) = x**4 - 14*x**3 + 60* x**2 -70*x$')
plt.scatter(xmin1, f(xmin1), color='red', zorder=5, label=f'Minimizer at x = {xmin1:.6f}')
plt.title('Function Plot with Minimizer (Golden Search)')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True)
plt.show()
```
